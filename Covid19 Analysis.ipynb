## Calculating the first derivative of the curve

```python
#show the peak point infections reached in a 24 hours spot
plt.figure()
covid_dataset_sum.loc['France'].diff().plot()
#print the max infection rate occured in 24 hours
print('Max infection rate happened in a 24 \
hours slot is in France is: ', covid_dataset_sum.loc['France'].diff().max())
```

### Retrive max infection rate for all countries 

```python
found = list()
for c in list_countries:
    found.append(covid_dataset_sum.loc[c].diff().max())
```

```python
#Add the a new column to the dataset sum
covid_dataset_sum['max_infection_rate'] = found
```

```python
covid_dataset_sum.head()
```

```python
covid_dataset_sum.columns
```

```python
#create a new dataframe with only the max_infection_rate
covid_inf_rate = pd.DataFrame(covid_dataset_sum['max_infection_rate'])
```

```python
#check the newly created dataframe
covid_inf_rate.head()
```

```python
#let nalyse the data from world_happiness datasabset
worldwide_happiness.head()
```

```python
worldwide_happiness.head()
```

```python
#let create a list with the important columns for our analysis
columns = ['GDP per capita', 'Social support', 'Healthy life expectancy', 'Perceptions of corruption']
```

```python
#keep just the importants columns
worldwide_happiness = worldwide_happiness[columns]
```

```python
worldwide_happiness.head()
```

```python
#let merge our two dataframe
data = covid_inf_rate.merge(worldwide_happiness, how = 'inner', 
                            left_on = covid_inf_rate.index, right_on = worldwide_happiness.index)
```

```python
#check the newly dataFrame
data.head()
```

```python
#optional => set index to the countries' name
data = data.set_index('key_0')
```

## Let's do the same analysis with the death dataset due to Covid19

```python
#check the dataset
covid_deaths.head()
```

```python
#remove useless columns
df = covid_deaths.drop(['Lat', 'Long'], axis = 1, inplace = True)
```

```python
covid_deaths.head()
```

```python
#create a new dataframe with the sum of deaths for every country
covid_deaths_sum = covid_deaths.groupby('Country/Region').sum()
```

```python
covid_deaths_sum.head()
```

```python
#SImple way to compare the number of death in some countries
plt.figure()
covid_deaths_sum.loc['France'].plot()
covid_deaths_sum.loc['Germany'].plot()
covid_deaths_sum.loc['Poland'].plot()
covid_deaths_sum.loc['China'].plot()
plt.legend()
```

```python
#retrieve the max rate of death occured in 24 hours spot
covid_deaths_sum.loc['France'].diff().max()
```

```python
#create a list of all max death rate occured in a 24 hour spot
found2 = []
list_countries_2 = list(covid_deaths_sum.index)
for c in list_countries_2:
    found2.append(covid_deaths_sum.loc[c].diff().max())
```

```python
#create a new column with the rate compilled
covid_deaths_sum['max_death_rate'] = found2
```

```python
covid_deaths_sum.head()
```

```python
#create a new dataFrame with the only required column
covid_death_rate = pd.DataFrame(covid_deaths_sum['max_death_rate'])
```

```python
covid_death_rate.head()
```

```python
#join the death DT with the world happiness DT
data_d = covid_death_rate.merge(worldwide_happiness, how = 'inner', 
                                left_on = covid_death_rate.index, right_on = worldwide_happiness.index)
```

```python
#check the newly created DT
data_d.head()
```

```python
#Optional => set index to country names
data_d = data_d.set_index('key_0')
```

## Plotting results

##### From the graphs below, we will find that the number of infected people is greater in developped countries

### Plotting GDP per capita vs max death rate

```python
plt.figure()
sns.scatterplot(data_d['GDP per capita'], np.log(data_d['max_death_rate']))
```

### Plotting Healthy life expectancy per capita vs max death rate

```python
plt.figure()
sns.scatterplot(data_d['Healthy life expectancy'], np.log(data_d['max_death_rate']))
```
